```python
import h5py
import numpy as np
```

```python
pi0 = h5py.File("../dataset/Pi0Escan_1_10_DownSampled.h5", "r")
pi0_ECAL = pi0.get("ECAL")
pi0_HCAL = pi0.get("HCAL")
#pi0_energy = pi0.get("energy")
print(pi0_ECAL.shape)
print(pi0_HCAL.shape)
```

```python
pion = h5py.File("../dataset/ChPiEscan_1_1_DownSampled.h5", "r")
pion_ECAL = pion.get("ECAL")
pion_HCAL = pion.get("HCAL")
#pion_energy = pion.get("energy")
print(pion_ECAL.shape)
print(pion_HCAL.shape)
```

```python
ele = h5py.File("../dataset/EleEscan_1_1_DownSampled.h5", "r")
ele_ECAL = ele.get("ECAL")
ele_HCAL = ele.get("HCAL")
#pi_energy = pi.get("energy")
print(ele_ECAL.shape)
print(ele_HCAL.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import LogNorm
```

```python
SUM_Image = np.sum(ele_ECAL, axis = 0)
plt.imshow(SUM_Image/float(ele_ECAL.shape[0]), origin='lower',norm=LogNorm(vmin=0.01))
plt.colorbar()
plt.title("Average Ele ECAL", fontsize=15)
plt.xlabel("x cell", fontsize=15)
plt.ylabel("y cell", fontsize=15)
plt.show()
SUM_Image = np.sum(ele_HCAL, axis = 0)
plt.imshow(SUM_Image/float(ele_HCAL.shape[0]), origin='lower',norm=LogNorm(vmin=0.01))
plt.colorbar()
plt.title("Average Ele HCAL", fontsize=15)
plt.xlabel("x cell", fontsize=15)
plt.ylabel("y cell", fontsize=15)
plt.show()
```
